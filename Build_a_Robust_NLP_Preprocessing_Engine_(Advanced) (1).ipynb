```python
#Task 2

import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return []

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters/emojis
    text = re.sub(r"[^\w\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words (<=2 except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return tokens
```

```python
#Task 3

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    cleaned = " ".join(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-"*50)

```

```python
# Task 4

def token_analytics(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0

    return total, unique, avg_len


```

```python
# Task 4

for sentence in test_sentences:
    tokens = preprocess_text(sentence)

    total, unique, avg = token_analytics(tokens)

    print("Sentence:", sentence)
    print("Total:", total)
    print("Unique:", unique)
    print("Average Length:", round(avg, 2))
    print("-"*50)
```

```python
# Task 5

from collections import Counter

all_tokens = []

for sentence in test_sentences:
    all_tokens.extend(preprocess_text(sentence))

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

```python
# Task 6

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_sentences.append(" ".join(tokens))
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
# Task 6

result = full_pipeline(test_sentences)
print(result)
```

```python
# Task 7

def preprocess_text(text):

    if not isinstance(text, str):
        return []
    if text.strip() == "":
        return []
    import re

    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return tokens
```
